```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
matplotlib.rcParams['figure.figsize'] = (12,12)
```

```python
df = pd.read_csv('raw data/movies.csv')
```

```python
df.head()
```

```python
# Check for missing data percentage per column

df.isnull().mean()*100
```

```python
df.dtypes
```

```python
# Clean data types 

# df['budget'] = df['budget'].astype('int64')
# df['gross'] = df['gross'].astype('int64')

#error due to NAN/Null values
```

```python
df['Year_Correct'] = df['released'].astype(str).apply(lambda st: st[st.find(",")+1:st.find("(")-1])
```

```python
df['Year_Correct'] = df['Year_Correct'].astype(str).apply(lambda st: st[st.find(" ")+1:])
```

```python
df.dtypes
```

```python
df.isnull().mean()*100
```

```python
df = df.sort_values(by=['gross'], inplace= False, ascending= False)
```

```python
# drop duplicated

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plot Gross vs Budget

plt.scatter(x=df['budget'], y = df['gross'])
plt.title('Gross Earning vs Budget')
plt.ylabel('Gross Earnings')
plt.xlabel('Film Budget')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data =df, scatter_kws={"color":"red"}, line_kws={'color': 'blue'})
```

```python
# Correlation
df.corr() #pearson (default), kendall, spearman
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
plt.show()
```

```python
# # # Look at Company, convert text of company name to numeric value

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr  = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
